```python
print(f"\n{'='*70}")
print('Univariate analysis on each column')
print(f"{'='*70}\n")

def categorical_analysis(column):
  column.value_counts().plot(kind='bar')

print(f"\n{'='*70}")
print('Univariate analysis on Gender column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Gender']))

print(f"\n{'='*70}")
print('Univariate analysis on [Item Purchased] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Item Purchased']))

print(f"\n{'='*70}")
print('Univariate analysis on [Size] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Size']))

print(f"\n{'='*70}")
print('Univariate analysis on [Color] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Color']))

print(f"\n{'='*70}")
print('Univariate analysis on [Season] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Season']))

print(f"\n{'='*70}")
print('Univariate analysis on [Subscription Status] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Subscription Status']))

print(f"\n{'='*70}")
print('Univariate analysis on [Shipping Type] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Shipping Type']))

print(f"\n{'='*70}")
print('Univariate analysis on [Discount Applied] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Discount Applied']))

print(f"\n{'='*70}")
print('Univariate analysis on [Promo Code Used] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Promo Code Used']))

print(f"\n{'='*70}")
print('Univariate analysis on [Payment Method] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Payment Method']))

print(f"\n{'='*70}")
print('Univariate analysis on [Frequency of Purchases] column')
print(f"{'='*70}\n")

plt.show(categorical_analysis(df['Frequency of Purchases']))
```

```python
print(f"\n{'='*90}")
print('Extracting only numeric columns from the Dataset and excluding Customer ID ')
print(f"{'='*90}\n")
numerical_col=[col for col in df.columns if df[col].dtype != 'object'][1:]
print(numerical_col)

print(f"{'='*90}\n")
print(f'Skewness analysis')
print(f"{'='*90}\n")
print(df[numerical_col].skew())

for col in numerical_col:
  print(f"{'='*90}\n")
  print(f'Boxplot of the {col}')
  print(f"{'='*90}\n")
  plt.show(sns.boxplot(df[col]))

  print(f"{'='*90}\n")
  print(f'histogram + kde  of the {col}')
  print(f"{'='*90}\n")
  plt.show(sns.histplot(df[col],kde=True))
```

```python
print(f"\n{'='*90}")
print('Multivariate Analysis')
print(f"{'='*90}\n")

plt.show(sns.pairplot(df))

plt.figure(figsize=(15,6))
plt.show(sns.boxplot(y=df['Item Purchased'],x=df['Purchase Amount (USD)']))



```
